```python
import json

# Open the JSON file and read its content
with open("response.json", "r",encoding='utf-8') as file:
    data = json.load(file)  # This reads the file and parses the JSON content
print(data)
```

```python
import pandas as pd
```

```python
title = []
price = []
brand = []
id = []
for i in range(len(data['data'])):
    title.append(data['data'][i]['name'])
    id.append(data['data'][i]['id'])
    brand.append(data['data'][i]['brand_name'])
    price.append(data['data'][i]['price'])
```

```python
df = pd.DataFrame(list(zip(id,title,brand,price)),columns=['id','title','brand','price'])
```

```python
df.groupby('brand')['price'].count().reset_index().sort_values('price',ascending=False).reset_index(drop=True)
```
